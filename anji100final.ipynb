```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
df = pd.read_csv('youtube.csv')
print(df.head())
print(df.info())
print(df.isnull().sum())
```

```python
#Create a line plot
value_counts = df['Country'].value_counts()
plt.figure(figsize=(12, 8))
plt.plot(value_counts.index, value_counts.values, marker='o', color='green', linewidth=2, label='Channel Count')
plt.title('Country-wise Channel Distribution', fontsize=14, color='green')
plt.xlabel('Country', fontsize=14, color='blue')
plt.ylabel('Number of Channels', fontsize=14, color='red')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=14)
plt.savefig('country_distribution_line_chart.png', dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()


#violin_plot_subscribers
plt.figure(figsize=(10, 6))
sns.violinplot(x='Category', y='Subscribers (millions)', data=df, palette='muted')
plt.title('Distribution of Subscribers by Category', fontsize=14)
plt.xlabel('Category', fontsize=14, color='green')
plt.ylabel('Subscribers (millions)', fontsize=14, color='red')
plt.xticks(rotation=45)
plt.savefig('violin_plot_subscribers.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# brand vs. non-brand channels
def plot_brand_vs_nonbrand_pie_chart(data, save_path='brand_vs_nonbrand_pie_chart.png'):
    """
    Plots a pie chart comparing brand vs. non-brand channels.

    Parameters:
    - data (DataFrame): The dataset containing the 'Brand channel' column.
    - save_path (str): File path to save the figure. Default is 'brand_vs_nonbrand_pie_chart.png'.
    """
    brand_channel_counts = data['Brand channel'].value_counts()
    plt.figure(figsize=(12, 8))
    plt.pie(
        brand_channel_counts, 
        labels=brand_channel_counts.index, 
        autopct='%1.2f%%', 
        startangle=80
    )
    plt.title('Comparison of Brand vs Non-Brand Channels', fontsize=14, color='Orange')
    plt.legend(fontsize=12)
    plt.savefig('Comparison of Brand vs Non-Brand Channels.png', dpi=300, bbox_inches='tight')
    plt.show()
   

plot_brand_vs_nonbrand_pie_chart(df)
```

```python
# Histogram with KDE
plt.figure(figsize=(15,8)) 
sns.histplot(df['Primary language'], bins=6, kde=True, color='purple', edgecolor='black', alpha=0.7, label='Histogram with KDE')
plt.title("Distribution of Primary Languages with KDE", fontsize=17, color='blue')
plt.xlabel("Primary Language", fontsize=17, color='red')
plt.ylabel("Number of Channels", fontsize=17, color='green')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title_fontsize=12, fontsize=14, loc='upper right')
plt.savefig('primary_language_distribution.png', dpi=300, bbox_inches='tight')
plt.show()
```
